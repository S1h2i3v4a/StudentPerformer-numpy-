```python
import numpy as np
```

```python
# Load only numeric marks (columns 1 to 5), skip header row, skip name column
marks = np.genfromtxt('sample_data.csv', delimiter=',', skip_header=1, usecols=(1,2,3,4,5))

# Load only names column separately
names = np.genfromtxt('sample_data.csv', delimiter=',', skip_header=1, usecols=0, dtype=str)

print("Names:", names)
print("Marks:\n", marks)
print("Shape:", marks.shape)
```

```python
stu = np.average(marks, axis=1)  # Calculate average marks for each student (row-wise)
 

# Single highest/lowest mark in entire dataset (not average)
highest_mark = np.max(marks)
lowest_mark = np.min(marks)

# Find WHICH student and subject
row, col = np.unravel_index(np.argmax(marks), marks.shape)
print(f"Highest mark: {highest_mark} by {names[row]} in subject{col+1}")

row, col = np.unravel_index(np.argmin(marks), marks.shape)
print(f"Lowest mark: {lowest_mark} by {names[row]} in subject{col+1}")
```

```python
conditions = [stu >= 90, stu >= 80, stu >= 70, stu >= 60]
choices = ['A', 'B', 'C', 'D']
grades = np.select(conditions, choices, default='F')

for i in range(len(names)):
    print(f"{names[i]} -> avg: {stu[i]:.2f}, grade: {grades[i]}")
```

```python
# argsort default ascending deta hai, isliye reverse karna padega
sorted_indices = np.argsort(stu)[::-1]   # descending order ke indices

# Top 3
print("Top 3 Performers:")
for i in sorted_indices[:3]:
    print(f"{names[i]} -> {stu[i]:.2f}")

# Full ranking
print("\nFull Ranking:")
for rank, i in enumerate(sorted_indices, start=1):
    print(f"Rank {rank}: {names[i]} -> {stu[i]:.2f}")
```

```python
sub_avg = np.average(marks, axis=0)  # Calculate average marks for each subject (column-wise)
sub_max = np.argmax(marks, axis=0)  # For each subject (column), find the index of the student with highest mark
print("\nTop Performers in Each Subject:")
for j, idx in enumerate(sub_max):
    print(f"Subject {j+1}: {names[idx]} with mark {marks[idx][j]}")
sub_min = np.argmin(marks, axis=0)  # For each subject (column), find the index of the student with lowest mark
print("\nLowest Performers in Each Subject:")
for j, idx in enumerate(sub_min):
    print(f"Subject {j+1}: {names[idx]} with mark {marks[idx][j]}")
sub_indices = np.argsort(sub_avg)[::-1]  # Sort subjects by average marks in descending order
print(f"Subjects toughest: Subject {sub_indices[-1] + 1}")
print(f"Subjects easiest: Subject {sub_indices[0] + 1}")
```
